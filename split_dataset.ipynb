SPLIT DATASET (BAS and ATM)

```python
BAS_DATASET_PATH = '/data/dengxiaolong/airway/BAS/Data/'
ATM_DATASET_PATH = '/data/dengxiaolong/airway/ATM/train/'
RANDOM_SEED = 42
```

```python
import os
import pickle
import random
```

BAS SPLIT

```python
image_dir = os.path.join(BAS_DATASET_PATH, 'image')
image_files = [f for f in os.listdir(image_dir)]
image_files.sort()
print(image_files)
```

```python
# split 7:1:2
total_size = len(image_files)
train_end = int(total_size * 0.7)
val_end = train_end + int(total_size * 0.1)

random.seed(RANDOM_SEED)
random.shuffle(image_files)

train_images = image_files[:train_end]
val_images = image_files[train_end:val_end]
test_images = image_files[val_end:]
data_split = {
    'train': train_images,
    'val': val_images,
    'test': test_images,
}
data_split
```

```python
with open('BAS_dataset_split.pickle', 'wb') as f:
    pickle.dump(data_split, f)
```
